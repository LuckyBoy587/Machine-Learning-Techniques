<b>Numpy Arrays</b>

```python
import numpy as np
array1=np.array([1,2,3,4,5])
print(array1)
```

```python
array1.shape
```

```python
a=np.array([[1,2,3,4],[5,6,7,8]])
print(a)
a.shape
```

```python
b=np.array([[1,2,3,4],[5,6,7,8]])
print(b)
b.shape
```

```python
#create a numpy array of zeroes
x=np.zeros((4,5))
print(x)
```

```python
y=np.ones((3,3))
print(y)
```

```python
z=np.full((5,4),5)
print(z)
```

```python
#create an identity matrix
i1=np.eye(4)
print(i1)
```

```python
#create a numpy array with random values
rand1=np.random.random((3,4))
print(rand1)
```

```python
#random values array within a specific range
rand2=np.random.randint(10,100,(3,5))
print(rand2)
```

```python
#Array of evenly spaced values
arr=np.linspace(10,30,5)
print(arr)
```

```python
#Array of evenly spaced values
arr2=np.arange(10,30,5)
print(arr2)
```

```python
#find array dimension
print(rand2.shape)
```

```python
#number of dimensions
print(rand2.ndim)
```

```python
#Number of elements in an array
print(rand2.size)
```

```python
#checking the datatype of the values in the array
print(rand2.dtype)
```

```python
#Math operations performed on np array
A=np.random.randint(0,10,(3,3))
B=np.random.randint(10,20,(3,3))
print(A)
print(B)
print(A+B) #print(np.add(A,B)
```

```python
print(A-B)#print(np.sub(A,B)
```

```python
print(np.multiply(A,B)) #print(A*B)
```

```python
print(np.divide(A,B)) #print(A/B)
```

```python
#Transpose
A1=np.random.randint(0,5,(2,3))
print(A1)
print(A1.shape)
```

```python
trans=np.transpose(A1)
print(trans)
print(trans.shape)
```

```python
#Reshaping the array
B1=A1.reshape(3,2)
print(B1)
```
