<span style="font-size:20px;"><b>Vector Operations in Python</b></span>

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#plotting a Vector
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

<span style="font-size:20px;"><b>Addition of two Vectors</b></span>

```python
vector1=np.asarray([0,0,2,3])
vector2=np.asarray([0,0,3,-2])
vectsum=vector1+vector2
print(vectsum)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='y')
plt.quiver(0,0,5,1,scale_units='xy',angles='xy',scale=1,color='r')

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

<span style="font-size:20px;"><b>Subtraction of two vectors</b></span>

```python
vector1=np.asarray([0,0,2,3])
vector2=np.asarray([0,0,3,-2])
sub=vector1-vector2
print(sub)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='y')
plt.quiver(0,0,-1,5,scale_units='xy',angles='xy',scale=1,color='r')

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

<span style="font-size:20px;"><b>Multipying a vector by a scaler</b></span>

```python
vector1=np.asarray([0,0,2,3])
vector2=2*vector1
print(vector2)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,4,6,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

<span style="font-size:20px;"><b>Dot product of 2 vectors</b></span>

```python
a=np.array([2,3])
b=np.array([4,4])
adotb=np.dot(a,b)
print(adotb)
c=np.array([40,20,35])
d=np.array([53,24,68])
cdotd=np.dot(c,d)
print(cdotd)
```

<span style="font-size:20px;"><b>cross product</b></span>

```python
across=np.cross(a,b)
ccrossd=np.cross(c,d)
print(across)
print(ccrossd)
```

<span style="font-size:20px;"><b>Projection</b></span><br>



```python
a=np.array([2,5])
v=np.array([8,-6])
```

```python
#Magnitude of v vector
magv=np.sqrt(sum(v**2))
proja=(np.dot(a,v)/magv**2)*v
print(magv)
print(proja)
```
